# Feature Engineering

## Handiling missing value

```python
# HANDLING CATAGORICAL MISSING VALUES
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
for feature in feature_nan:
    print(feature, np.round(dataset[feature].isnull().mean(),4), ' % missing value')
    
```

```python
def replace_cat_features(dataset, feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data
```

```python
dataset = replace_cat_features(dataset, feature_nan)
```

```python
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
## Handing missing numerical values
numerical_feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']
for feature in numerical_feature_nan:
    print(feature, np.round(dataset[feature].isnull().mean(),4), ' % missing value')
```

```python
for feature in numerical_feature_nan:
    median = dataset[feature].median()
    
    dataset[feature + 'nan'] = np.where(dataset[feature].isnull(),0,1)
    dataset[feature].fillna(median, inplace = True)
dataset[numerical_feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold']- dataset[feature]
```

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

```python
## Handing the rare catagorical feature
```

```python
catogorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == "O"]
```

```python
catagorical_features
```

```python
for feature in catagorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.1].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature], 'RareVar' )
```

```python
dataset.head()
```

# Feature Scaling

```python
dataset.head()
```

```python
for feature in catagorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
