```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.shape
```

```python
dataset.head()
```

## finding the columns that have null values

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() >1 ]
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4), ' % missing value')
```

## since there are many missing values, we need to find the relation between sale price and the columns which have missing values.

```python
for feature in features_with_na:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

## Numerical Variables

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype!='O']
```

```python
len(numerical_features)
```

```python
dataset[numerical_features].head()
```

## Temporal variable

```python
temporal_features = [feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]
temporal_features
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median of SalePrice')
plt.title('')
```

```python
for feature in temporal_features:
    if feature != 'YrSold':
        data = dataset.copy()
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.show()
```

## Discrete Variables

```python
discrete_Variables = [feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in temporal_features + ['Id']]
len(discrete_Variables)
```

```python
discrete_Variables
```

```python
dataset[discrete_Variables].head()
```

```python
for feature in discrete_Variables:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

## Continuous Variables

```python
continuous_variable = [feature for feature in numerical_features if feature not in discrete_Variables + temporal_features + ['Id']]
len(continuous_variable)
```

```python
continuous_variable
```

```python
for feature in continuous_variable:
    data = dataset.copy()
    data[feature].hist(bins = 20)
    plt.xlabel(feature)
    plt.show()
```

```python
# we will be using logarithimic transformation

for feature in continuous_variable:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.show()
        data[feature].hist()
        plt.show()
```

## Outliers

```python
for feature in continuous_variable:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.show()
```
